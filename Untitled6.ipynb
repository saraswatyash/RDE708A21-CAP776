```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("covid19.xlsx")
print(df.columns)
```

```python
df["Nationality"]=df["Nationality"].fillna("NIA") #to fill all the values with nia
nation=df["Nationality"].value_counts()
nation["India"]+=nation["Indian"]
nation=nation.drop(labels=["NIA","Indian"])#to drop indian and NIA values
nation.plot.pie(figsize=[15,15],title="Infected Nationalities",autopct="%1.1f%%",
                    explode=[0.1,0.5,0.5,0.5,0.5,0.5,0.5,0.5])
plt.show()
```

```python
ds=df["Detected State"].value_counts()#to count every distinct values 
ds.plot.bar(figsize=(15,15),title="No. of cases per state",color='#8661B8')
plt.show()
```

```python
state_gen=df.groupby(["Detected State","Gender"])["Gender"].count()#groups the value by state and shows according to gender
state_gen.unstack().plot.bar(figsize=[15,15],stacked=True,color=["#FF00FF","#104E8B"])
plt.show()
```

```python
df["Age Bracket"]=df["Age Bracket"].fillna("NIA")
age=df["Age Bracket"].value_counts()
age=age.drop(labels="NIA")
age.plot.barh(figsize=(15,15),color='r')
plt.show()
```
